```python
# Importing liabraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Creating Data frame

```python
df = pd.read_csv("Zomato data .csv")
```

```python
df
```

```python
df.info()
```

# Converting the Data type of rating Column - rate

```python
def handlerate(value):
    value = str(value).split('/')
    value = value[0];
    return float(value)

df['rate'] = df['rate'].apply(handlerate)

df
```

```python
df.isnull().sum()
```

# Type of Restaurants

```python
df.head(5)
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("Type of Restaurants")
```

# Conclusion - Majority of the restaurants falls under Dinning Category

```python
df.head(5)
```

```python
group_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': group_data})

plt.figure(figsize=(10, 6))  # Adjust the figure size
plt.plot(result, color="green", marker='o')
plt.plot(result, color="red") 
plt.ylabel('Votes', color="red", fontsize=20)  # Correcting ylabel and setting font size
plt.xlabel('Type of Restaurants', fontsize=20)  # Adding xlabel with a set font size
plt.title('Votes by Type', fontsize=24) # adding a title
plt.xticks(rotation=45)  # Rotating x-axis labels for better readability
plt.grid(True)

plt.show()
```

# Conclusion- Dinning restaurants has received maximum votes from users

```python
plt.hist(df['rate'], bins=10)
plt.title("Rating Distribution")
plt.show()
```

# Conclusion- The Majority of Rating lies Between 3.5 to 4 and Highest number of rating fall in 4

# Average order Spending by Users

```python
spending = df['approx_cost(for two people)']
plt.figure(figsize=(10, 6))
sns.countplot(x=spending, palette='viridis')  
plt.xticks(rotation=45)
plt.title('Distribution of Approximate Costs for Two People', fontsize=16)
plt.xlabel('Approximate Cost (for two people)', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
spending = df['approx_cost(for two people)']

top_5_spending = spending.value_counts().nlargest(5).index

filtered_df = df[df['approx_cost(for two people)'].isin(top_5_spending)]

plt.figure(figsize=(10, 6))
sns.countplot(x='approx_cost(for two people)', data=filtered_df, order=top_5_spending)
plt.title('Top 5 Most Common Approximate Costs for Two People')
plt.xlabel('Approximate Cost (for two people)')
plt.ylabel('Count')
plt.xticks(rotation=45)  
plt.show()
```

Conclusion - Maximum ticket size is Rs. 950 but most number of orders Comes at Rs.300 Followed by 200,150,400 and 500

# Average Spending on each order

```python
description = df['approx_cost(for two people)'].describe()
average_cost = description['mean']
print(f"Average approximate cost for two people: {average_cost}")
```

# Conclusion- Average approximate cost for two people is 418

# Which Mode Received Maximum rating

```python
plt.figure(figsize=(6,6))
sns.boxenplot(x = 'online_order', y = 'rate', data=df)
```

# Conclusion - Online order Recived more good rating than offline

```python
df.head(5)
```

```python
pivot_table = df.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
plt.figure(figsize=(10, 6))  
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')  

plt.title("Heatmap of Online Orders by Type", fontsize=16)
plt.xlabel("Online Order", fontsize=14)
plt.ylabel("Listed in (Type)", fontsize=14)

plt.show()
```

Conclusion - Dining restaurants have most of the offline Orders, While cafes primarly receives Online orders. This Suggest that Clients preferes to Order Offline at restaurants but prefer online ordering at cafes.

### Project Outcomes Summary:

1. **Majority of Restaurant Types**:
   - The majority of restaurants fall under the **Dining category**. This indicates a strong customer preference for sit-down dining experiences where they can enjoy a more relaxed and social atmosphere.

2. **Customer Votes and Preferences**:
   - **Dining restaurants** have received the maximum number of votes from users. This suggests that customers highly favor these types of establishments over others, likely due to the quality of food, ambiance, and service they provide.

3. **Restaurant Ratings**:
   - The majority of restaurant ratings lie between **3.5 to 4 stars**, with the highest number of ratings being **4 stars**. This range indicates that overall customer satisfaction is quite high, reflecting positive dining experiences and good service standards.

4. **Order Value and Frequency**:
   - The maximum ticket size observed is **₹950**, but most orders are placed at a lower price point. The most common order values are around **₹300**, followed by **₹200**, **₹150**, **₹400**, and **₹500**. This pattern suggests that customers frequently opt for affordable meal options, with a significant number making purchases at the ₹300 price range.

5. **Average Spending for Couples**:
   - The average approximate cost for two people dining out is **₹418**. This figure provides insight into the spending behavior of couples, indicating that they prefer moderately priced meal options.

6. **Mode of Ordering and Ratings**:
   - **Online orders** have received more positive ratings compared to offline orders. Customers appreciate the convenience, ease, and perhaps additional features such as discounts and variety offered by online platforms.

7. **Offline vs. Online Order Preferences**:
   - **Dining restaurants** see the majority of offline orders, while **cafes** primarily receive online orders. This trend suggests that customers prefer a traditional in-person dining experience when visiting restaurants but opt for the convenience of online ordering when choosing cafes or quick bites.

These insights provide valuable information for strategic decision-making, such as focusing on enhancing the online ordering experience for cafes and developing targeted promotions for dining restaurants to boost offline orders. Understanding customer preferences and spending behavior will allow Zomato to tailor its offerings more effectively to meet customer needs and expectations.
